```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
logins = pd.read_json("/content/drive/MyDrive/ultimate_tech_takehome/logins.json")
```

```python
logins.head()
```

```python
logins.tail()
```

```python
logins.info()
```

```python
logins_agg = logins.set_index("login_time")
logins_agg['count'] = 1
logins_agg = logins_agg.resample("15min").sum()
```

```python
logins_agg.info()
```

```python
logins_agg.plot()
plt.show()
```

```python
logins_daily = logins.set_index('login_time')
logins_daily['count'] = 1 
logins_daily = logins_daily.resample("D").sum()
logins_daily.plot()
```
